# text cleaning

```python

#removing puncuations 

import re

REPLACE_NO_SPACE = re.compile("[.;:!\'?,\"()\[\]]")
REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def preprocess_reviews(reviews):
    reviews = [REPLACE_NO_SPACE.sub("", line.lower()) for line in reviews]
    reviews = [REPLACE_WITH_SPACE.sub(" ", line) for line in reviews]
    
    return reviews

reviews_clean = preprocess_reviews(clothing['Review Text'])
# reviews_test_clean = preprocess_reviews(reviews_test)
reviews_clean
```

```python
from nltk.corpus import stopwords

english_stop_words = stopwords.words('english')
# stop_words=['in','of','at','a','the','and']
def remove_stop_words(corpus):
    removed_stop_words = []
    for review in corpus:
        removed_stop_words.append(
            ' '.join([word for word in review.split() 
                      if word not in english_stop_words])
        )
    return removed_stop_words

reviews_clean = remove_stop_words(reviews_clean)
reviews_clean
```

### Implementing Potter Stemmer

```python


def get_stemmed_text(corpus):
    from nltk.stem.porter import PorterStemmer
    stemmer = PorterStemmer()
    return [' '.join([stemmer.stem(word) for word in review.split()]) for review in corpus]

stemmed_reviews = get_stemmed_text(reviews_clean)
stemmed_reviews
```
